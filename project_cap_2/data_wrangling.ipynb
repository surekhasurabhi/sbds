```python
missing_summary = df.isnull().sum()
missing_summary = missing_summary[missing_summary > 0]
print(missing_summary.sort_values(ascending=False))
```

```python
df = df.dropna(axis=1, thresh=len(df)*0.8)
```

```python
numeric_cols = df.select_dtypes(include=['float', 'int']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

```python
df['Metro'] = df['Metro'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
id_vars = ['City Code', 'City', 'Metro', 'County', 'State', 'Population Rank']

value_vars = [col for col in df.columns if col not in id_vars]

df_long = pd.melt(
    df1,
    id_vars=id_vars,
    value_vars=value_vars,
    var_name='Date',
    value_name='RentPrice'
)
```

```python
df_long
```

```python
df_long = df_long.dropna(subset=['RentPrice'])
df_long.head()
```

```python
city_example = 'San Francisco'
city_df = df_long[df_long['City'] == city_example]

plt.figure(figsize=(12,6))
plt.plot(city_df['Date'], city_df['RentPrice'], marker='o', linestyle='-')
plt.title(f'Rent Price Trend in {city_example}', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price per Sqft', fontsize=12)
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_long.to_csv('zillow_rent_cleaned.csv', index=False)
```
