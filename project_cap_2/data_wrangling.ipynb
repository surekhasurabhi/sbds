```python
def cap_outliers_iqr(group):
    Q1 = group['RentPrice'].quantile(0.25)
    Q3 = group['RentPrice'].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    group['RentPrice'] = np.clip(group['RentPrice'], lower, upper)
    return group
```

```python
df_long_capped = (
    df_long
    .groupby('City', group_keys=False)
    .apply(cap_outliers_iqr)
)
```

```python
print("Original RentPrice range:")
print(df_long['RentPrice'].describe())

print("\nCapped RentPrice range:")
print(df_long_capped['RentPrice'].describe())
```

```python
city_example = 'San Francisco'
city_df = df_long_capped[df_long_capped['City'] == city_example]

plt.figure(figsize=(12,6))
plt.plot(city_df['Date'], city_df['RentPrice'], marker='o')
plt.title(f'Rent Price Trend in {city_example} (Outliers Capped)', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Rent Price')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```
